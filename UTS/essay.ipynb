3. Taxonomy dalam pengerapan machine learning adalah pengelompokan berbagai macam teknik machine learning berdasarkan karakteristiknya. Taxonomy ini dapat digunakan untuk memudahkan pemahaman dan penggunaan machine learning.

Berikut adalah beberapa macam taxonomy dalam pengerapan machine learning:

Berdasarkan jenis data yang digunakan

Berdasarkan jenis data yang digunakan, machine learning dapat dibagi menjadi dua kategori, yaitu:

Machine learning dengan data numerik. Machine learning dengan data numerik menggunakan data yang hanya terdiri dari angka. Data numerik ini dapat berupa data kontinu, seperti tinggi badan, berat badan, atau suhu, atau data diskrit, seperti kode pos atau nilai ujian.
Machine learning dengan data non-numerik. Machine learning dengan data non-numerik menggunakan data yang tidak hanya terdiri dari angka. Data non-numerik ini dapat berupa data teks, data gambar, atau data audio.
Berdasarkan jenis tugas yang diselesaikan

Berdasarkan jenis tugas yang diselesaikan, machine learning dapat dibagi menjadi tiga kategori, yaitu:

Supervised learning. Supervised learning adalah jenis machine learning di mana komputer dilatih pada kumpulan data yang berisi label atau target. Komputer kemudian menggunakan data ini untuk mengembangkan model yang dapat digunakan untuk memprediksi hasil baru.
Unsupervised learning. Unsupervised learning adalah jenis machine learning di mana komputer tidak dilatih pada kumpulan data yang berisi label atau target. Komputer kemudian menggunakan data ini untuk menemukan pola atau struktur dalam data.
Reinforcement learning. Reinforcement learning adalah jenis machine learning di mana komputer dilatih untuk mengambil tindakan yang menghasilkan hasil yang diinginkan. Komputer kemudian menggunakan data ini untuk mengembangkan model yang dapat digunakan untuk mengambil tindakan yang optimal dalam situasi yang tidak diketahui.
Berdasarkan jenis algoritme yang digunakan

Berdasarkan jenis algoritme yang digunakan, machine learning dapat dibagi menjadi berbagai kategori, antara lain:

Algoritme klasifikasi. Algoritme klasifikasi digunakan untuk mengklasifikasikan objek atau peristiwa ke dalam kategori yang berbeda. Contoh algoritme klasifikasi adalah logistic regression, decision tree, dan support vector machine.
Algoritme regresi. Algoritme regresi digunakan untuk memprediksi nilai kontinu dari suatu variabel. Contoh algoritme regresi adalah linear regression, polynomial regression, dan logistic regression.
Algoritme clustering. Algoritme clustering digunakan untuk mengelompokkan objek atau peristiwa yang memiliki kesamaan. Contoh algoritme clustering adalah k-means, hierarchical clustering, dan density-based clustering.
Algoritme association rule learning. Algoritme association rule learning digunakan untuk menemukan aturan-aturan yang menunjukkan hubungan antara dua atau lebih variabel. Contoh algoritme association rule learning adalah Apriori dan Eclat.
Algoritme natural language processing. Algoritme natural language processing digunakan untuk memahami dan menanggapi bahasa manusia. Contoh algoritme natural language processing adalah part-of-speech tagging, named entity recognition, dan machine translation.
Taxonomy dalam pengerapan machine learning dapat membantu dalam memahami dan menggunakan machine learning. Dengan memahami berbagai macam teknik machine learning, kita dapat memilih teknik yang tepat untuk menyelesaikan masalah yang dihadapi.
